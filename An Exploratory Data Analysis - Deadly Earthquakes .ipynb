## Data Visualization  

I have added some visualization graphs and charts to show the data in a more understanable way!

#### 1. Countplot

```python
plt.figure(figsize = (15,6))
sns.countplot(df['MMI'], data = df, palette = 'hls')
plt.xticks(rotation = 90)
plt.show()
```

#### 2. Pie Chart

```python
plt.figure(figsize= (30,20))
plt.pie(df['MMI'].value_counts(), labels = df['MMI'].value_counts().index, autopct = '%1.1f%%', textprops= {'fontsize' : 25, 'color' : 'black', 'weight' : 'normal', 'family' : 'cursive'})

hfont = {'fontname': 'serif', 'weight' : 'black'}
plt.title('MMI', size=20 , **hfont)
plt.show()
```

####  3. Histplot

Here , I have used 'Magnitude' column for visualisation

```python
plt.figure(figsize= (15,6))
sns.histplot(df['Magnitude'], kde = True, palette = 'hls') # kde is used to show the continous line kind of standard deviation line
plt.xticks(rotation = 90)
plt.show()
```

Now, I have used 'Depth' column for visualisation

```python
plt.figure(figsize= (15,6))
sns.histplot(df['Depth'], kde= True , palette = 'hls')
plt.xticks(rotation = 90)
plt.show()
```

Now, I have used 'MMI' column for visualisation

```python
plt.figure(figsize= (15,6))
sns.histplot(df['MMI'], kde= True , palette = 'hls')
plt.xticks(rotation = 90)
plt.show()
```

#### 4. Scatterplot

```python
plt.figure(figsize = (15,6))
sns.scatterplot(x= df['Magnitude'] , y= df['Depth'], palette = 'hls')
plt.show()
```

To compare from one column to two different axis, use 'hue' function. This means the below chart is accroding to MMI to two different axis's information which is Magnitude and depth.

```python
plt.figure(figsize= (15,6))
sns.scatterplot (x= df['Magnitude'], y= df['Depth'], hue = df['MMI'] , palette = 'hls')
plt.show()
```

#### 5. Lineplot

```python
plt.figure(figsize = (15,6))
sns.lineplot(x= df['Magnitude'], y= df['Depth'],  palette = 'hls')
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.lineplot(x= df['Magnitude'], y= df['Depth'], hue = df['MMI'],  palette = 'hls')
plt.show()
```
