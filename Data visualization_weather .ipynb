```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('Test.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
#weather.info() is used to know data types and data values
```

```python
#sns.barplot is used to have bar graphs
```

```python
sns.barplot(weather['humidity'],weather['temperature'])
```

```python
#2nd one is displot. It is used for continusouly data representation.
```

```python
sns.displot(weather['humidity'])
```

```python
#3rd one is Bivariant in which we can use more than 2 columns
```

```python
sns.jointplot(weather['humidity'],weather['temperature'])
```

```python
sns.jointplot(weather['humidity'],weather['temperature'],kind="hex")
```

```python
#hex is used for having hexagon than dotes in the graph
```

```python
sns.jointplot(weather['humidity'],weather['temperature'],kind="kde")
```

```python
#kde means kernel density estimates
```

```python
#if we want to show data as pair wise, we can use pairplot.
```

```python
sns.pairplot(weather[['humidity','temperature','air_pollution_index']])
```

```python
#for categorical difference. we can use stripplot function
```

```python
sns.stripplot(weather['weather_type'],weather['humidity'])
```

```python
sns.boxplot(weather['humidity'],weather['temperature'],hue=weather['weather_type'])
```

```python
#Here hue is used as weather type, so the data is being displayed or being categorized from the weather type's data.
```

```python
sns.barplot(weather['humidity'],weather['temperature'],hue="weather_type",data=weather)
```
